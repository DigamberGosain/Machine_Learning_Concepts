# Making pivot Tables

```python
pdf = pd.read_csv("Sales_Funnel_CRM.csv")
```

```python
pdf.columns.values
```

```python
license = pdf[["Company",'Product','Licenses']]
```

```python
pd.pivot(data=license, index="Company",columns="Product", values="Licenses")
```

```python
pd.pivot_table(pdf, index= "Company", aggfunc='sum', values=['Licenses', 'Sale Price'], margins=True)
```

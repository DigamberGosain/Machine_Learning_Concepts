# <span style = "color:purple;font-family:bremen bd bt"> Distribution Plots - Understanding plot types</span>

### Distribution plots displays a sigle continous feature and help visualize properties such as deviation and average values

#### There are 3 main distributiob plot types:
###### Rug plot , Histogram, KDE plot (KDE - Kernel Density estimation)


## Rug plot: it merely adds a dash or tick line for every single value along with x axis. The y-axis doesn't really have a meaning. But we can see outliers easily.

## Histogram is based on bins created for a continous feature based on counts in that bins, it also inclused the y-axis as it tells the count occurance.

## Seaborn allows us to add KDE plot curve on top of histogrm which shows what a continous probability function will look like
### KDE stands for <span style = "color:red">kernel Density Estimation</span>. It is a method of estimation a probability density function of a random varibale

```python
#Rug Plot
# y-axis doesn't have any importance


sns.rugplot(x = 'salary', data = df);
```

```python
#Rug Plot
# We can increase the tick size as well
sns.rugplot(x = 'salary', data = df, height= 0.5)
```

```python
# we can change figures based on matplotlib like figure size
```

```python
#DO not used the DIST plot
# Distribution plot

sns.displot(data = df, x = 'salary')
```

```python
# Distribution plot
# We can increase a decreas the bin size
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
# We can change the style as well like backround
#sns.set(style = 'darkgrid')#, "whitegrid","white","dark", "ticks". sns.set will change the style for all entire page
# if we want to change it comment out sns.set command and then go to kerner>restart and run all
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
# for editing color and other style changes are same as Matplotliba nd present in sns.displot (color, hue etc.)
#**kwargs are avaibale as per matplotlib
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', edgecolor = 'blue', lw = 4)
```

```python
#Histplot
sns.histplot(data = df, x ='salary')
```

```python
# Dis and histplotare same 
# with Diplot we can use KDE as a parmeter
# even we can bring all plots together even rug plot
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', edgecolor = 'blue', lw = 4, kde = True, rug = True);
```

```python
# if we only need KDE plot
sns.kdeplot(data = df, x = 'salary');
```

# Categorical Plots- Distribution within Categories

## What should we do if we want to compare ditributions across categories?
### We will seperate put each category then create the distribution visualization
## Distribution with in categories plots
#### boxplot()
#### Violinplot()
#### Swarmplot()
#### Boxenplot(Letter-Value Plot)

##### Most famous plot is Box plot

#### The Boxplot displays the distributions of a continous varibales. It does this though the use of quartiles
#### We can actually create a box plot per category


## The Violin plot plays a similar role as the box plot, but it display the density using KDE


## The Swarm plot is very simple and simply shows all the date point in the distribution
#### For Very large datasets, it won't show all the points but will display the general distributoin of them. Its's one dot for every data point

## The boxenplot(letter-value plot) is a relatively new plot, It's mainly designed as and explansion upon the normal box plot.


### With boxen-plot it uses a system of letter 0 values we can use multiple quantiles instead of strictly using quartiles

# <span style = "color:#03520F;font-family:bremen bd bt"> Comparison Plots - Understanding plot types</span>


## Comparison plots are essentially 2-dimensional plots
### The two main plots tupes are:
#### jointplot()
#### pairplot()

## jointplot()
### We can map histograms to each feature of a scatterplt to clarify the distributions within each feature.
### We can also adjust the scatterplot to be a hex plot or a 2d KDF plot

#### <span style = "color:red;font-family:bremen bd bt"> Note: Hexagons are dark the more points falls into their area </span>

# pairplot()
### The pairplot is a quick way to compare all numerical columns in a DataFrame
### It automatically creates histogram for each column and a scatter plot comparison between all possible combination of columns

#### <span style = "color:red;font-family:bremen bd bt"> WARNING: pairplot can be cpu and ram intensive for large DataFrame with many columns. It is a good idea to first filter down onlt the columns you are interested in </span>

```python
df1.head()
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score');
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score', kind = 'hex');# id we don't want points
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score', kind = 'scatter', alpha = 0.3);# or we can add alpha for points on top
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score', kind = 'hist', alpha = 0.5);# or 'hist' 
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score', kind = 'kde', shade = True);# or to show KDE we use kind - KDE
```

```python
sns.jointplot(data = df1, x = 'math score', y = 'reading score', hue = 'gender');# add a hue for another level
```

# <span style = "color:#03520F;font-family:bremen bd bt"> Seaborn Grid plot</span>


## Seaborn grid calls uses Matplotlib subplots() to automatically create a grid based off a categorical column
## Instead of passing ina specific number of cols or rows for the subplots, we can simply pass in the name of the columns and seaborn will automatically map the subplots grid

## Many of seaborn's built- in plot calls are running on top of this grid system
## Direclty call the grid system allows users to heavily customize plots.

### under grid plot, first plot is catplot()

```python
sns.catplot(data = df1, x ='gender', y = 'math score', kind = 'box');
```

```python
sns.catplot(data = df1, x ='gender', y = 'math score', kind = 'box', col = 'lunch', row = 'test preparation course');
```

```python
g = sns.PairGrid(df1, hue = 'gender')
g = g.map_upper(sns.scatterplot)# for upper diagonal
g = g.map_diag(sns.histplot)# or sns.kdeplot
g = g.map_lower(sns.kdeplot)
```

```python
# To add legend
g = sns.PairGrid(df1, hue = 'gender')
g = g.map_upper(sns.scatterplot)# for upper diagonal
g = g.map_diag(sns.histplot)# or sns.kdeplot
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```

# <span style = "color:brown;font-family:bremen bd bt"> Seaborn Matrix plot</span>


# Matrix plots are the visual equivalent of displaying a pivot table

## The matrix plot displays all the data passed in , visualizing all the numeric values in a DataFrame.

#### <span style = "color:red;font-family:bremen bd bt"> Note: Not every DataFrame is a valid choice for a matrix plot such as a heatmap
</span>

### The tow main matrix plot types are
### Headtmap()
#### Visually displays the distribution of cell value with a color mapping

### Clustermap()
#### Same visual as headmap , but first conducts hirerachial clustering to reorganize data into groups


## <span style = "color:red;font-family:bremen bd bt"> Note: A headmap should ideally have all cells be in the same units , so the color mapping make sense accross the entire DataFrame
</span>

### Seaborn also comes with the ability to automatically cluster simial groupings

## Headmaps

##### In the above example life expantancy is in different units than the other features, so it's color is different and not matching the otherrs. In this scenaro only take the columns which are in the same units

```python
# coloumn clustering sometime doesn;t make sense, so we remove it 
sns.clustermap(df2.drop('Life expectancy', axis = 1), linewidths=0.5, 
               annot = True, cmap = 'viridis', col_cluster=False);
```

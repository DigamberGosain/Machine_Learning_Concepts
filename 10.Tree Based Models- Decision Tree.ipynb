# <span style = "color:#02424a;fornt-family:Helvetica"> Decision Tree - Understanding Gini Impurity

## <span style = "color:#02424a;fornt-family:Helvetica"> Gini Impurity is a mathematical measurement of how "pure" the information in a data set is.</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> In regards to classification, we can think of this as a measurement of class uniformity.

# <span style = "color:#02424a;fornt-family:Helvetica"> Gini Impurity for classification is defined as following:
* For a set of classes C for a given dataset Q: 
    $\large{G}{(}{Q}{)}{ = }\sum_{{c}\in{C}}{p}_{c}{(}{1}{ - }{p}_{c}{)}$
* Where ${p}_{c}$ is probability of Class c.
    $\large{p}_{c}=\frac{1}{{N}_{Q}}\sum_{{x}\in{Q}}{1}{(}{y}_{class}{ = }{c}{)}$

# <span style = "color:#02424a;fornt-family:Helvetica">Let's Look at a example: We have four data points</span>
### <span style = "color:red;fornt-family:Helvetica"> 2 belong to red class</span>
### <span style = "color:blue;fornt-family:Helvetica"> 2 belong to blue class</span>

### <span style = "color:red;fornt-family:Helvetica"> Class Red = (2/4)(1-2/4) = 0.25</span>
### <span style = "color:blue;fornt-family:Helvetica"> Class Blue = (2/4)(1-2/4) = 0.25</span>

#### <span style = "color:#02424a;fornt-family:Helvetica"> In general we have to perform this calculation for each clases in the dataset </span>

### <span style = "color:#02424a;fornt-family:Helvetica"> So, as per the formula we have sum both classes probability to get Gini Impurity</span>
## <span style = "color:#02424a;fornt-family:Helvetica"> Gini impurity = 0.25 + 0.25 = 0.5

## <span style = "color:#02424a;fornt-family:Helvetica"> We can see from the curve at 0.5 we have maximum gini impurity

## <span style = "color:#02424a;fornt-family:Helvetica"> When Gini Impurity is 0 it means data is "pure"(no impurity). Which also means there is only one class in the data

## <span style = "color:#02424a;fornt-family:Helvetica"> Where does Gini Impurity fits in decision tree</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> If the goal of a decision tree is to separate out classes, we can use gini impurity to decide on data split values.</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> We want to minimize the gini impurity at leaf nodes.</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Minimized impurity ar the leaf node means we are seperating classes effectively!
   

# <span style = "color:#02424a;fornt-family:Helvetica"> Constructing Decision Trees with Gini Impurity

```python
sp = {"X-URL Link":["Yes","Yes","No","No","No","No","Yes"],"Y-Spam":["Yes","Yes","No","No","Yes","No","No"]}
df = pd.DataFrame(sp)
df
```

#### <span style = "color:#02424a;fornt-family:Helvetica"> The above is an example data, and we will construct Decision tree based on two fearure X and y whether or not email is spam

### <span style = "color:#02424a;fornt-family:Helvetica"> Here we have only one X feature to use for a node.</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> With this feature we have only two possibilities Yes or No. So if we have to split the node we have either "Yes" or "NO".</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> So if we split URL we will see how many times when URL is attached in the email it was Spam, and not attached </span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> Whether Spam being "Yes" or "No" when URL present we have 2: Yes and 1: No</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> Whether Spam being "Yes" or "No" when URL present we have 1: Yes and 2: No</span>

### <span style = "color:#02424a;fornt-family:Helvetica"> Now we will calculate Gini Impurity for the entire URL feature</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> First Leave node calculation for two classes "Yes" & "No":
*For Yes Class $\large{(}\frac{2}{3}{)}{(}{1}{ - }\frac{2}{3}{)}$</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> Now we will add No Class to Yes:
* Yes + No $\large{(}\frac{2}{3}{)}{(}{1}{ - }\frac{2}{3}{)} { + }{(} \frac{1}{3}{)}{(}{1}{ - }\frac{1}{3}{)}$</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> Left Leaf Gini = 0.44</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> We will repeat the same calculation for Right leaf node:
* $\large{(}\frac{1}{4}{)}{(}{1}{ - }\frac{1}{4}{)} { + }{(} \frac{3}{4}{)}{(}{1}{ - }\frac{3}{4}{)}$</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> Right Leaf Gini = 0.375
    
    


## <span style = "color:#02424a;fornt-family:Helvetica"> Now we calculate gini impurity of URL feature.</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Weighted Average of both:
    * Left Leaf Gini = 0.44
    * Right Leaf Gini = 0.375</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Total number of emails: (2+1)(1+3) = 7</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Our weighted average will be based on how many emails are on each nodes:
* Left node emails : 3
* Right Node emails : 4 </span>
### <span style = "color:#02424a;fornt-family:Helvetica"> So, Our Wighted average calcuation will be:
* $\large{(}\frac{3}{7}{)}$ * ${0.44}{ + }    {(}\frac{4}{7}{)}$ * ${0.375}$</span>
#### <span style = "color:#02424a;fornt-family:Helvetica"> By calcualting above we get Gini Impurity for URL= 0.403


## <span style = "color:#02424a;fornt-family:Helvetica"> Now lets take the Gini Impurity for:
* Continous numeric features
* Multi- categorical features
* Choosing a root node feature

## <span style = "color:#02424a;fornt-family:Helvetica"> Continous numeric feature

## <span style = "color:#02424a;fornt-family:Helvetica"> The first step is sort the continous variable

### <span style = "color:#02424a;fornt-family:Helvetica"> In the above example as we can see after sorting we take the average of two numbers on the row for e.g 7 + 12 = 19 and then average of that number i.e 19/2 = 9.5</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Our first node became age $\le$ 9.5</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> If N is less than equal to 9.5 then we have 1 no and 0 yes and if N is greater than 9.5 than we have 3 yes and 3 no then our calcuation for gini will be following.
*(1/1 * (1-1/1) = 0 for yes and 0 as well for no on the left node which means 0 + 0</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> for Right node we have of N is greater than 9.5 then our calcualtion is (3/6 * 1-(3/6)) = 0.25 for yes and same for No which is 0.25+ 0.25=0.5</span>
### <span style = "color:#02424a;fornt-family:Helvetica"> Final Gini calculation is
(1/7)(0)+(6/7)*0.5 = 0.429

## <span style = "color:#02424a;fornt-family:Helvetica"> This way we will calculate Gini for each split based on the average of two rows and average with lowest impurity is used for splits

## <span style = "color:#02424a;fornt-family:Helvetica"> By Choosing the feature with the lowest resulting gini impurity in its leaf nodes, we are choosing the feature that best splits the date into "pure" classes.

### <span style = "color:#02424a;fornt-family:Helvetica"> Also, by using gini impurity as mesurement of the effectiveness of a node split, we can perform feature selection by mandating and impurity threshold for an additional feature based split to occur.

### <span style = "color:#02424a;fornt-family:Helvetica"> In terms of modelling decision tree, so far Scikit_learn doesn't support multi categories being passed in as strings, for now we have to enter dummy varibales to the model.

```python
X = pd.get_dummies(df.drop("species", axis = 1),drop_first=True)
```

```python
y = df["species"]
```

## <span style = "color:#02424a;fornt-family:Helvetica"> We don't need to scale the features as in decision tree it directly compare the entire dataset on a single feature. We will not use multiple feature at the same time.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## <span style = "color:#02424a;fornt-family:Helvetica"> Decision Tree - creating model

```python
from sklearn.tree import DecisionTreeClassifier
```

## <span style = "color:#02424a;fornt-family:Helvetica">Following are the default paramenters for DecisionTreeClassifier:
DecisionTreeClassifier(*, criterion='gini', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, class_weight=None, presort='deprecated', ccp_alpha=0.0)
 |  

## <span style = "color:#02424a;fornt-family:Helvetica"> First models with default parameter

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
print(classification_report(y_test,base_preds))
```

```python
plot_confusion_matrix(model, X_test,y_test)
```

### <span style = "color:#02424a;fornt-family:Helvetica"> The good part of Decision tree it is very easily interpretable by human and it comes with lot of usefull attribute once it s fitted

## <span style = "color:#02424a;fornt-family:Helvetica"> Scikit-Learn also comes with option to visualise tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi =200)
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi =200)
    plot_tree(model, feature_names=X.columns, filled=True);
```

## <span style = "color:#02424a;fornt-family:Helvetica"> We will use differnt model by changing the hyperparameter , first one is by reducing the dept of the tree as the base model if we can see form above has 5 levels.
 

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

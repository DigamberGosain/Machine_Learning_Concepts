# <span style = "color:#03689e;font-family:helvetica"> Pricipal Component Analysis

### <span style = "color:#03689e;font-family:helvetica"> So Far in unsupervised learning we have focused on clustering techniques, which seek to discover labels on feature data that has no historical labels.</span>

### <span style = "color:#03689e;font-family:helvetica"> Now we will shift toward unsupervised algorithm that focus on dimention reduction

### <span style = "color:#03689e;font-family:helvetica"> Motivation behind Dimension Reduction:
* Imagine a dataset with 30+ Features, how would you understand key features?
* Visualizations and Data Analysis have limitations when the number of feautre dimentiosn increases

### <span style = "color:#03689e;font-family:helvetica"> There maily two outcomes we look from Dimensionality Reduction:
* Understand which feature describes the most variance in the data set.
* Aid human understanding of large feature sets, especially through visualization.

### <span style = "color:red;font-family:helvetica"> Important Note:
* Dimensionality Reduction algorithms such as PCA do not simply choose a subset of the existing features.
* They create new dimenstional components that are combinations of proportions of the existing features

### <span style = "color:#03689e;font-family:helvetica"> Principal Component Analysis Outcomes:
* Reduce number of dimensions in data Which means transform data into lower dimension
* Show which feature explain the most variance in the data, and this is unlabeled or unsupervised data techinique. 

### <span style = "color:#03689e;font-family:helvetica"> Dimension Reduction: 
* Helps us to visualize and understand complex data sets.
* Can also act as simpler data set for training data for machine learning algorithms.
    * Reduce dimensions then trin ML algorithm on samller data set
* Help reduce N features to a desired smaller set of componenets through a transformation.
* It does not simply select a subset of features.

### <span style = "color:#03689e;font-family:helvetica"> Variance explained:
* We have often seen that certain features are more important or have moore explanotary power than other features.
* For example, size of the house is more important than the color of the house when explaining the price of a house for sale.
* This idea of more important features is easy to understand when we can directyl correlate features to a know label. But what about unlabled data?
* What meassurement can we use to determine feature importance or variance explained by each feature?
    

### <span style = "color:#03689e;font-family:helvetica"> Well we can measure the proportion to which each feature accounts for dispersion in the data set.

#### <span style = "color:#03689e;font-family:helvetica"> In two feature space show above PCA will try to look where the variance is best explained as in the above diagram, so we can see PCA 2nd dimension explains the variabce well. It will be marked as Principal component 1 and its nice about PCA it keeps creates components by count.

### <span style = "color:#03689e;font-family:helvetica"> Principal component is a linear combination of original features.</span>
### <span style = "color:#03689e;font-family:helvetica"> The more variance the origianl feature accounts for, the more influence it has over the principal components.

### <span style = "color:#03689e;font-family:helvetica">So with the above diagram we went from 2 features down to 1 pricncipal component.</span>

### <span style = "color:#03689e;font-family:helvetica"> The single principal component can "explain" some percentage of the original data, for example 90% of variance explained by principal component and variance explained means some sort of measurement of dispersion.

### <span style = "color:#03689e;font-family:helvetica"> Variance explained:
* 100% of the variance in the data is explained by all the origianl features.
* We trade off some of the explained variance for less dimensions.
* This can be significant savings for data sets with many dimensions, but only a few strong features.

# <span style = "color:#03689e;font-family:helvetica"> PCA Intuition

### <span style = "color:#03689e;font-family:helvetica">Principan Component Analysis operated by creating a new set of dimensions(the principal componenets) that are normalized linear combinations of the origianl features.
$\large{Z}_{1}{ = } \phi_{11}{X}_{1}{ + } \phi_{21}{X}_{2}{ +.....+}\phi_{p1}{X}_{p}$

### <span style = "color:#03689e;font-family:helvetica"> This means if we have found some PCA , it will be a linear combination of number of features like in above equation

### <span style = "color:#03689e;font-family:helvetica"> So with two dimension feature space we can figure out PCA "Z1" and that will be a linear combination for feature of x1 and x2 like  $\large{Z}_{1}{ = } \phi_{11}{X}_{1}{ + } \phi_{21}{X}_{2}$ With PCA we can reduce from two dimension  to 1 dimension.

## <span style = "color:#03689e;font-family:helvetica"> Now question comes how do we calculate these components?

### <span style = "color:#03689e;font-family:helvetica"> First step in the PCA is to Standardize the data:
* While doing the standardization we are actually shifting the mean of the entire dataset towards zero

### <span style = "color:#03689e;font-family:helvetica"> Also by doing this we haven't changed the relationship of the data but the mean is shifted toward zero

## <span style = "color:#03689e;font-family:helvetica"> Step 2 is to Calculate covariance matrix of the data

### <span style = "color:#03689e;font-family:helvetica"> After doing linear transformation vector becomes Eigen Vector: which represet Directional Information which is same after transfomration but change happens with Magnitude information

### <span style = "color:#03689e;font-family:helvetica"> Now for every Eigen vector there is going to be an Orthogonal EigenVector which means just perpendicular to the EigenVector. EigenVector is just linear transformation

### <span style = "color:#03689e;font-family:helvetica"> We use covariance matrix to apply linear transformation to the origianl dataset and that results in Pricipal component analysis and it will look like below 

$\large\left[\substack{{Var(X)}\\{Cov(X,Y)}} \substack{{Cov(X,Y)}\\{Var(Y)}}\right]$

### <span style = "color:#03689e;font-family:helvetica"> After this transformation we will discover eigen vector and so we can use eigen value to calcualte the magnitude of these vectors which would also corresponds to the variance explained

### <span style = "color:#03689e;font-family:helvetica"> So we will have PCA as EigenVectors and EigenValues which measures variance explained

### <span style = "color:#03689e;font-family:helvetica"> PCA steps in summary:
* Get original data
* calculate Covariance Matrix
* Calculate EigenVectors
* Sort EigenVectors by EigenValue
* Choose N Largest EigenValues
* Project Original data onto EigenVectors

# <span style = "color:#03689e;font-family:helvetica"> Principal Component Analysis Manual Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cancer_tumor_data_features.csv")
```

```python
df.head()
```

### <span style = "color:#03689e;font-family:helvetica"> We can see there are 30 columns which is very hard to understand as we can see below we can't make out anything from headmap as well

```python
sns.heatmap(df)
```

### <span style = "color:#03689e;font-family:helvetica"> We will start with Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis = 0)
```

### <span style = "color:#03689e;font-family:helvetica"> We can see after scaling means for each point is close to 0

### <span style = "color:#03689e;font-family:helvetica"> Now step two grab covariance matrix

```python
covariance_matrix = np.cov(scaled_X, rowvar= False)## because we want cov across features
```

### <span style = "color:#03689e;font-family:helvetica"> Numpy comes with built linear algebra library which calculates the EigenValue and EigenVector automatically

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors.shape
```

### <span style = "color:#03689e;font-family:helvetica"> Next step is how many components would you like to take form this

### <span style = "color:#03689e;font-family:helvetica"> So we have moved from N feaure space to N PC component

### <span style = "color:#03689e;font-family:helvetica"> We will start with 30 feature space, sort eigeen value and choose top 2 starting to visualize

```python
num_component = 2
```

### We will use np.argsort for sorting eigenvalues as  argsort goint to sort everything and bring indcies for array

```python
np.argsort(eigen_values)
```

#### we will reverse this so that we can go from large values to small

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_component]
```

### Sorted_key still referes to indicies so we will do following 

```python
eigen_values,eigen_vectors = eigen_values[sorted_key], eigen_vectors[:,sorted_key]
```

### We still need to take original data and project it onto that eigenvector

### Dot product is a projection method to one vector to another

```python
principal_components = np.dot(scaled_X,eigen_vectors)
```

#### This step is taking our original data and projecting it to eigne vectors, scaled_X has 30 columns and we are projecting it down to 2 eigenvectors or 2 principal components sorted by top two values

```python
principal_components
```

```python
plt.scatter(principal_components[:,0], principal_components[:,1])
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dict = load_breast_cancer()
```

```python
cancer_dict.keys()
```

```python
print(cancer_dict["target"])
```

### This data actually had tartget column which has 2 classes and what we want to see here that even after reducing the data from 30 features to 2 pricipal components, can we still classify them correclty

```python
plt.scatter(principal_components[:,0], principal_components[:,1],c = cancer_dict["target"])
```

#### And we can see there is a perfect classfication even with two features

# <span style = "color:#03689e;font-family:helvetica"> PCA with Scikit_Learn

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

### <span style = "color:#03689e;font-family:helvetica"> class PCA(sklearn.decomposition._base._BasePCA)PCA(n_components=None, *, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)

### <span style = "color:#03689e;font-family:helvetica"> With Sklearn its supereasy we just to enter the number of components and rest we can leave default

```python
pca_model = PCA(n_components=2)
```

```python
pca_model.fit(scaled_X)### calculated the eigen values eigen vector
```

```python
pc_results = pca_model.transform(scaled_X) ## Projection on scaled data
```

```python
plt.scatter(pc_results[:,0],pc_results[:,1], c = cancer_dict["target"])
```

```python
pca_model.components_
```

### <span style = "color:#03689e;font-family:helvetica"> It explains the direction of the variance in the date and it is sorted based on the explained variance. We can visualize with heat map

```python
df_components = pd.DataFrame(pca_model.components_,index = ["PC1","PC2"], columns= df.columns)
```

```python
df_components
```

```python
plt.figure(figsize=(20,3))
sns.heatmap(df_components,annot = True)
```

### <span style = "color:#03689e;font-family:helvetica"> But much more important is explained variance

```python
pca_model.explained_variance_ratio_
```

### <span style = "color:#03689e;font-family:helvetica"> First Principal components contains 44% of variacne explained in our data, second one explains 19% of variance. So with just 2 components are able to explain almost 63% of variance in the dataset

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    
    pca = PCA(n_components= n)
    pca.fit(scaled_X)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explained_variance)
plt.xlabel("Num of Component")
plt.ylabel('Variance_explained')
```

# <span style = "color:#500463;font-family:helvetica">Random Forests

### <span style = "color:#500463;font-family:helvetica">Random Forests have the ability to greatly increase the perfromance based on expanding ideas from the Decision Tree.</span>
### <span style = "color:#500463;font-family:helvetica">Random Forests are know as ensemble learners, since they rely on ensemble of models(multiple decision trees).

# <span style = "color:#500463;font-family:helvetica"> Motivation behind Random Forest

## <span style = "color:#500463;font-family:helvetica"> With Decision Tree if we recall, we could try adjusting rules such as:
* Splitting Criterion(Infromation Gain)
* Minimum Gini Impurity Decrease
* Setting Depth Limits
* Limits on number of terminal leaf nodes

## <span style = "color:#500463;font-family:helvetica"> However for decison tree Root node feature has huge influence over tree, Root node will aways be the same for the same set of rules.</span>
## <span style = "color:#500463;font-family:helvetica">These are the kind of issues we keep in mind while building out a random forest

### <span style = "color:#500463;font-family:helvetica">For all the above mentioned single set of rules there is only one set of decision tree possible.</span>
### <span style = "color:#500463;font-family:helvetica"> What we need to realise here that even with all these adjusted hyperparameters, the single decision tree we get is still limited:
* Single feature for root node.
* Splitting criteria can lead to some feature not being used.
* Potential for overfitting to data.
    

## <span style = "color:#500463;font-family:helvetica"> The key idea behind the random forest is to create subsets of randomly picked featurs at each potential split</span>

## <span style = "color:#500463;font-family:helvetica"> Now let's say we have data and we create random subset of data and we have to find a feature within the subset of the feature as rood node based on the information criterion.</span>
### <span style = "color:#500463;font-family:helvetica">We repeat the process for the different subset, by doing this we explore various features</span>
### <span style = "color:#500463;font-family:helvetica">The question is how to get predictions out of many models. In classificaiton task We will tally up the votes of each tree for e.d 3 differnt trees 1st tree outpout is 0 , second tree out put is 0 and third tree output is 1, we will go with 0 as majority is with 0</span>

## <span style = "color:#500463;font-family:helvetica"> For continous feature Random forest trees can output the continous labels. In this case we will average out the predicted output.</span>

### <span style = "color:#500463;font-family:helvetica">for e.g., Tree1 predcied 10, Tree2 predcited 12 and Tree 3 predicrted 11, so average for (10,13,11) = 11.3

## <span style = "color:#500463;font-family:helvetica"> Adding a random feature subspace constraint could allow trees to grow much deeper without causing overfitting.</span>

# <span style = "color:#500463;font-family:helvetica"> Random Forrest- Number of estimators and features in Subsets

## <span style = "color:#500463;font-family:helvetica"> Random forrest estimators means n_number of decison trees
## <span style = "color:#500463;font-family:helvetica"> Number of Estimators
* Intuitevely, we know the more decison trees, the more opportunities to learn from a avariety of feature subset combinations.


## <span style = "color:#500463;font-family:helvetica"> How to choose number of trees?
* Reasonable default value is 100
* Suggestion are between 64 to 128 trees.
* Good way to find is do grid search of trees.
* We can plot error versus number of trees (similar to elbow method of KNN)
    * We should notice for diminishing error reduction after some N Trees.

## <span style = "color:#500463;font-family:helvetica"> Just like elbow method at some number of tree the error rate doesn't go futher down</span>
### <span style = "color:#500463;font-family:helvetica"> After certain number of trees, two things can occur:
* Different random selections don't reveal any more information.
    * Tree becomes highly correlated.
* Different random selections are simply duplicating trees that have already been created

### <span style = "color:#500463;font-family:helvetica"> Over all it allows us to be quite lenient in setting number of estimators hyperparameters,as overfitting is of minimal concern. 

# <span style = "color:#500463;font-family:helvetica"> Rnadom Forest - Selecting number of features

### <span style = "color:#500463;font-family:helvetica">Recall that if we some N number of features in a dataset, we are going to select a subset when we are splitting at a node</span>
### <span style = "color:#500463;font-family:helvetica"> So question is how many N we choose?</span>
#### <span style = "color:#500463;font-family:helvetica"> With original publications it was suggested subset of ${log}_{2}$(N +1) random features in subset given a set of N total features.

## <span style = "color:#500463;font-family:helvetica"> Based on the supervised learning Task, wether it is Calssification or regression trying to complete we may want to adjust our default N-features hyperparameters value for the number of features in a subset.</span>
### <span style = "color:#500463;font-family:helvetica"> current suggested convention is sqrt(N) in the subset given N features.</span>
### <span style = "color:#500463;font-family:helvetica"> And later suggestion indicated N/3 may be more suitable for regression taks, which is larger than sqrt(N).

### <span style = "color:#500463;font-family:helvetica"> Whether it is sqrt(N+1), sqrt(N) or N/3 all of them are findings of past research, but not a setting stone convection to follow, instead they can be used as starting point of grid search.Since it is very likely depending on the dataset we need to tweek the hyperparameters to optimise model perfromance
   

## <span style = "color:#500463;font-family:helvetica"> As a review of Hyperparameters discussed N-estimators and N-features:
* Number of estimators:
    * Start with 100 as default and feel free to grid search for higer value.
* Number of Features for Selection:
    * Start with sqrt(N), grid search for other possible values (N/3)

# <span style = "color:#500463;font-family:helvetica"> Random Forests - Bootstrap and Out-of-Bag Error

## <span style = "color:#500463;font-family:helvetica"> Bootstrap Samples:
We will trying to answre should we allow bootstrap samplling of each training subset of feature?

## <span style = "color:#500463;font-family:helvetica"> What is Bootstrapping?
* A term used to describe "random sampling with replacement".

### <span style = "color:#500463;font-family:helvetica"> We can see form the above on original data is sampled in 4 differnt way with differnt combinations but all are random and with replacement

## <span style = "color:#500463;font-family:helvetica"> Bootstrapping in Random Forest:
* For Each split we are randomly selecting a subset of feature.
* This random subset of features helps create more diverse trees that are not correlated to each other
* To further differentiate trees, we could bootstarp a slection of rows for each split.
* This resuls in two randomised training components:
    * Subset of feature used 
    * Bootstrapped rows of data</span>
### <span style = "color:#500463;font-family:helvetica">: So with bootstapping with each Decison tree in random forest when we are creating a split we grab a random subset of features, we also bootstrap rows of data along with it.
 

## <span style = "color:#500463;font-family:helvetica"> In layman terms when we are creating number of subset of features , within subset rows also selected randomly.

## <span style = "color:#500463;font-family:helvetica"> Why do do bootstrapping?
* Although, bootstrapping can be set to Fasle during training (as it is True by default)
* Botstrapping is yet another hyperparameter meant to reduce correlation between trees, since trees are then trained on different subsets of feature columns and data rows!. Reducing correlation helps our model to generalize well on the data hasn't seen before.
    

# <span style = "color:#500463;font-family:helvetica"> Out-Of-Bag Error

## <span style = "color:#500463;font-family:helvetica"> When we talk about out-of bag we are refering to bagging</span>
### <span style = "color:#500463;font-family:helvetica"> What is bagging?
* In Random Forest, we use bootstrapped data and then calculate a prediction based on the aggrgated prediction of the trees:
    * For classification: Most Voted Y class
    * Regression: Average of predicted Y's</span>
### <span style = "color:red;font-family:helvetica">Notice here we are using $\large\underline{b}$ootstrapped data and doing $\large\underline{ag}$gregated predictions.So, from underlined term the word baggin is created

## <span style = "color:#500463;font-family:helvetica"> If we performed bootstrapping when building out trees, this means that for certain trees, certain rows of data were used not for training as we are doing bootstrapped rows.

```python
import pandas as pd
data={"Orginal":[1,2,3,4,5,6],"B1":[1,1,3,4,6,2],"B2":[2,3,6,4,1,2],"B3":[2,3,4,2,6,2]}
df = pd.DataFrame(data)
df = df.set_index("Orginal")
df
```

## <span style = "color:#500463;font-family:helvetica"> If we look at the above data row 5 from original data is not used in any of the bootstrapped samples</span>
### <span style = "color:#500463;font-family:helvetica"> Which now means that row number 5 is not used in the training of trees, and due to this we can utilise this row as test data.</span>
### <span style = "color:#500463;font-family:helvetica"> These are know as Out of Bag Samples:
* Not used for constructing some trees.
* We could use these to get performance test metrics on treed that did not use thse rows!

## <span style = "color:#500463;font-family:helvetica">So with these out of bag rows we can use them can calculate y-ypred to get errosr or rather out of bag error.

## <span style = "color:#500463;font-family:helvetica"> Note that OOB score is a hyperparameter that doesn't really affect training process.</span>
### <span style = "color:#500463;font-family:helvetica"> It is seperate from bootstrapping, OOB score is an optional way of measuring performance, an alternative to using a standard train/test split, since bootstrapping naturally creating an unsued data during training.

#### <span style = "color:#500463;font-family:helvetica"> OOB is also limited to not using all the trees in the random forest, it can only be calcualted on tress that did not use the OOB data!</span>
#### <span style = "color:#500463;font-family:helvetica"> Due to not using the entire random forest, the default value of OOB score hyperparameter is set to FALSE in scikit-learn.

## <span style = "color:#500463;font-family:helvetica"> Next we will try Polynomial regression

```python
def run_model(model,X_train,y_train,X_test,y_test):
    # Fit the model
    model.fit(X_train,y_train)
    
    # Get Metrics
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    mae = mean_absolute_error(y_test,preds)
    print(f'MAE:{mae}')
    print(f'RMSE: {rmse}')
    
    
    #plot results MOdel signal Range
    signal_range = np.arange(1,100)
    signal_predict = model.predict(signal_range.reshape(-1,1))
    plt.figure(figsize=(10,5), dpi = 100)
    sns.scatterplot(x= "Signal",y = "Density",data = df, color = "black")
    plt.plot(signal_range, signal_predict)
```

```python
model = LinearRegression()
run_model(model,X_train,y_train,X_test,y_test)
```
